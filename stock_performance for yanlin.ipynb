```python
import numpy as np
import pandas as pd
data = pd.read_csv('涨跌幅.csv', delimiter=',', encoding='utf-8')
```

```python
data.head(5)
```

```python
#把数字中的逗号去掉
for i in range(data.shape[0]):
    data['总市值'][i] = float(data['总市值'][i].replace(',',''))
```

```python
#把数字中的逗号去掉
for i in range(data.shape[0]):
    if (type(data['区间涨跌幅'][i])!=type(1.0)):
        data['区间涨跌幅'][i] = float(data['区间涨跌幅'][i].replace(',',''))/100.0
    else:
        data['区间涨跌幅'][i]/100.0
```

```python
#对行业分组，然后对市值降序排序
group = data.groupby('行业名称', sort=False).apply(lambda x:x.sort_values('总市值', ascending=False)).reset_index(drop=True)
group
#对行业分组，然后对市值降序排序,每个行业取前5个
#data.groupby('行业名称', sort=False).apply(lambda x:x.sort_values('总市值', ascending=False)[0:5]).reset_index(drop=True)
```
